```python
from transformers import BitsAndBytesConfig
```

This code defines a BitsAndBytesConfig object, bnb_config, which configures the 4-bit quantization settings for the model. It then loads a pre-trained PaliGemmaForConditionalGeneration model with the specified 4-bit quantization configuration using the from_pretrained method.

The BitsAndBytesConfig object specifies:
load_in_4bit=True: Load the model in 4-bit precision.
- bnb_4bit_quant_type="nf4": Use numeric fixed-point 4-bit quantization.
bnb_4bit_compute_dtype=torch.bfloat16:
- Use 16-bit floating-point for compute data type.
- The device_map argument specifies which device (e.g., GPU or CPU) to use for the model. In this case, it's set to use device 0 (the default device).

By using 4-bit quantization, the model's memory usage and computational requirements are reduced, making it more efficient for deployment on devices with limited resources.



```python
# Define a BitsAndBytesConfig object for 4-bit quantization
bnb_config = BitsAndBytesConfig(
    # Load the model in 4-bit precision
    load_in_4bit=True,
    # Specify the 4-bit quantization type as "nf4" (numeric fixed-point 4-bit)
    bnb_4bit_quant_type="nf4",
    # Specify the compute data type as torch.bfloat16 (16-bit floating-point)
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Load a pre-trained PaliGemmaForConditionalGeneration model with 4-bit quantization
model = PaliGemmaForConditionalGeneration.from_pretrained(
    # Model ID
    model_id,
    # Quantization configuration
    quantization_config=bnb_config,
    # Device map (specify which device to use)
    device_map={"":0}
)
```

```python
model_id = "google/paligemma-3b-pt-224"
processor = PaliGemmaProcessor(model_id)
```

```python
device = "cuda"
```

```python
image_token = processor.tokenizer.convert_tokens_to_ids("<image>")
```

This custom collate function prepares a batch of examples for input to a model. It extracts the text and image data from each example, processes them using the processor (a tokenizer and image processor), and creates labels by cloning the input IDs and replacing padding and image tokens with -100. Finally, it converts the tokens to 16-bit floating-point and moves them to the specified device (e.g., GPU) before returning the batch.


```python
# Define a custom collate function for batching examples
def collate_fn(examples):
    # Extract the question and multiple choice answer from each example
    texts = ["answer " + example["question"] + "\n" + example['multiple_choice_answer'] for example in examples]

    # Extract the image from each example and convert to RGB format
    images = [example["image"].convert("RGB") for example in examples]

    # Process the text and images using the processor (tokenizer and image processor)
    tokens = processor(
        text=texts,
        images=images,
        # Return PyTorch tensors
        return_tensors="pt",
        # Pad to the longest sequence in the batch
        padding="longest",
        # Don't tokenize newline characters separately
        tokenize_newline_separately=False
    )

    # Create labels by cloning the input IDs
    labels = tokens["input_ids"].clone()

    # Replace padding token IDs with -100
    labels[labels == processor.tokenizer.pad_token_id] = -100

    # Replace image token IDs with -100
    labels[labels == image_token] = -100

    # Add the labels to the tokens dictionary
    tokens["labels"] = labels

    # Convert the tokens to torch.bfloat16 (16-bit floating-point) and move to the specified device (e.g., GPU)
    tokens = tokens.to(torch.bfloat16).to(device)

    # Return the processed batch of tokens
    return tokens
```

```python
from peft import get_peft_model, LoraConfig
```

This LoraConfig object defines the settings for applying LORA to a model. The r parameter sets the rank of the low-rank approximation, which controls the number of new parameters introduced by LORA. The target_modules list specifies which layers (modules) of the model to apply LORA to, and the task_type specifies the type of task being fine-tuned for (in this case, causal language modeling).



```python
# Define a LoraConfig object for Low-Rank Adaptation (LORA)
lora_config = LoraConfig(
    # Rank of the low-rank approximation (number of new parameters)
    r=8,
    # Specify the target modules (layers) to apply LORA to
    target_modules=[
        # Query projection layer
        "q_proj",
        # Value projection layer
        "v_proj",
        # Key projection layer
        "k_proj",
        # Value projection layer (again, possibly for a different attention head)
        "v_proj",
        # Gate projection layer (for output gating)
        "gate_proj",
        # Up projection layer (for output upsampling)
        "up_proj",
        # Down projection layer (for output downsampling)
        "down_proj"
    ],
    # Specify the task type (in this case, causal language modeling)
    task_type="CAUSAL_LM"
)
```

This code first loads a pre-trained PaliGemmaForConditionalGeneration model with 4-bit quantization settings using the from_pretrained method. Then, it applies LORA (Low-Rank Adaptation) to the model using the get_peft_model function, which modifies the model's weights to enable more efficient fine-tuning. Finally, it prints the trainable parameters of the modified model using the print_trainable_parameters method.


```python
# Load a pre-trained PaliGemmaForConditionalGeneration model with 4-bit quantization
model = PaliGemmaForConditionalGeneration.from_pretrained(
    # Model ID
    model_id,
    # Quantization configuration (4-bit quantization settings)
    quantization_config=bnb_config,
    # Device map (specify which device to use, in this case device 0)
    device_map={"":0}
)

# Apply LORA (Low-Rank Adaptation) to the model using the lora_config
model = get_peft_model(model, lora_config)

# Print the trainable parameters of the model
model.print_trainable_parameters()
```

```python
from transformers import TrainingArguments
```

```python
# Define the training arguments for the model
args = TrainingArguments(
    # Total number of training epochs
    num_train_epochs=2,
    # Batch size per device (GPU or CPU) for training
    per_device_train_batch_size=16,
    # Number of gradient accumulation steps before updating model weights
    gradient_accumulation_steps=4,
    # Number of warmup steps for the learning rate scheduler
    warmup_steps=2,
    # Initial learning rate for the optimizer
    learning_rate=2e-5,
    # Weight decay rate for regularization
    weight_decay=1e-6,
    # Beta 2 parameter for the Adam optimizer
    adam_beta2=0.999,
    # Logging steps for tracking training progress
    logging_steps=100,
    # Optimizer to use (in this case, AdamW with the Hugging Face variant)
    optim="adamw_hf",
    # Saving strategy (in this case, save at every 1000 steps)
    save_strategy="steps",
    # Save every 1000 steps
    save_steps=1000,
    # Limit the total number of saved checkpoints to 1
    save_total_limit=1,
    # Use 16-bit floating-point (bf16) for training
    bf16=True,
    # Report training metrics to TensorBoard
    report_to=["tensorboard"],
)
```

```python
from transformers import Trainer
```

```python
trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    data_collator=collate_fn
)
```
